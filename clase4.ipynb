```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)

histograma, _ = np.histogram(img.flatten(), 256, [0,256])

funcion_dist = histograma.cumsum()
hst_normalizado = (funcion_dist * histograma.max())/funcion_dist.max() #Resultante de la funcion integral

img_ecualizada = cv2.equalizeHist(img)


plt.figure(figsize=(15, 7))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(histograma, color = "black")
plt.title('Histograma original')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Normalizada')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hst_normalizado, color = "black")
plt.title('Histograma Normalizada')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('a.jpg', cv2.IMREAD_GRAYSCALE)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize= (8,8))

img_nueva = clahe.apply(img)

histograma, _ = np.histogram(img.flatten(), 256, [0,256])
histograma_nuevo, _ = np.histogram(img_nueva.flatten(), 256, [0,256])

plt.figure(figsize=(15, 7))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(histograma, color = "black")
plt.title('Histograma original')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(img_nueva, cmap='gray')
plt.title('Imagen Ecualizada')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(histograma_nuevo, color = "black")
plt.title('Histograma Ecualizada')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('cancer.jpg')

b, g, r = cv2.split(img)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
b_clahe = clahe.apply(b)
g_clahe = clahe.apply(g)
r_clahe = clahe.apply(r)

img_clahe = cv2.merge((b_clahe, g_clahe, r_clahe))

hist_b, _ = np.histogram(b_clahe.flatten(), 256, [0, 256])
hist_g, _ = np.histogram(g_clahe.flatten(), 256, [0, 256])
hist_r, _ = np.histogram(r_clahe.flatten(), 256, [0, 256])

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(hist_b, color='blue')
plt.plot(hist_g, color='green')
plt.plot(hist_r, color='red')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img_clahe, cv2.COLOR_BGR2RGB))
plt.title('Clahe')

plt.tight_layout()
plt.show()
```
