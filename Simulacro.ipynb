Ejercicio 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagenRecibida = cv2.imread('imagen1.jpg')
img_original = cv2.cvtColor(imagenRecibida, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 7))

#Se muestra la primera imagen
plt.subplot(1, 2, 1)
plt.imshow(img_original, cmap='gray')
plt.title('Original')
plt.axis('off')

#Se muestra la seegunda imagen recortada
plt.subplot(1, 2, 2)
plt.imshow(img_original[610:800,660:840,:], cmap='gray') #Se utiliza el axis con las coordenadas [610:800,660:840,:]
plt.title('Original')
plt.axis('off')

#Guardar la iamgen recortada
cv2.imwrite('imagenRecortada.jpg', cv2.cvtColor(img_original[610:800,660:840,:], cv2.COLOR_BGR2RGB))

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

linear_gradient = np.zeros((300, 300), dtype=np.uint8)
for x in range(300):
    linear_gradient[x, :] = int(255 * x / 300)

plt.figure(figsize=(10, 5))
plt.imshow(linear_gradient)
plt.title('Lineal')
plt.axis('off')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize= (8,8))

img_nueva = clahe.apply(linear_gradient)

histograma, _ = np.histogram(linear_gradient.flatten(), 256, [0,256])
histograma_nuevo, _ = np.histogram(img_nueva.flatten(), 256, [0,256])

plt.figure(figsize=(15, 7))

plt.subplot(2, 2, 1)
plt.imshow(linear_gradient)
plt.title('Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(histograma, color = "black")
plt.title('Histograma original')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(img_nueva)
plt.title('Imagen Ecualizada')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(histograma_nuevo, color = "black")
plt.title('Histograma Ecualizada')
plt.axis('off')

plt.show()
```

Ejercicio 5

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagenRecibida = cv2.imread('imagen1.jpg')
img_original = cv2.cvtColor(imagenRecibida, cv2.IMREAD_GRAYSCALE)

linear_gradient2 = np.zeros((900, 1350), dtype=np.uint8)

for x in range(900):
    linear_gradient2[x, :] = int(255 * x / 900)

fin_imagen = cv2.merge((img_original, linear_gradient2))

plt.figure(figsize=(15, 7))

plt.subplot(1, 2, 1)
plt.imshow(img_original)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(fin_imagen)
plt.title('Con degradado')

```
