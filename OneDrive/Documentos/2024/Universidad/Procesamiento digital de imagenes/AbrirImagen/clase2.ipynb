```python
import cv2
import matplotlib.pyplot as plt

def histograma(img):
    histo = cv2.calcHist([img],[0],None,[256],[0,256])

    img = cv2.imread('imagen.jpg', 0)
    _,umbral_bi = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) #Umbralizacion binaria 
    _,umbral_inv_bi = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #Umbralizacion binaria inversa

    umb_adap = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 25)
    umb_adap_gauss = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize=11, C=2)

    plt.figure(figsize=(20, 10))

    plt.subplot(231)
    plt.title('Imagen inicial')
    plt.imshow(img, cmap = "gray")
    plt.axis('off')

    plt.subplot(232)
    plt.plot(histo, color = "black")
    plt.title('Histograma')
    plt.xlabel('Intensidad pixel')
    plt.ylabel('Frecuencia pixel')

    plt.subplot(233)
    plt.title('Umbral binario')
    plt.imshow(umbral_bi, cmap='gray')
    plt.axis('off')

    plt.subplot(234)
    plt.title('Umbral binario invertido')
    plt.imshow(umbral_inv_bi, cmap='gray')
    plt.axis('off')

    plt.subplot(235)
    plt.title('Binario adaptativa')
    plt.imshow(umb_adap, cmap='gray')
    plt.axis('off')

    plt.subplot(236)
    plt.title('Binario adaptativo - Gauss')
    plt.imshow(umb_adap_gauss, cmap='gray')
    plt.axis('off')

    plt.show()

imagen = cv2.imread('imagen.jpg', 0)
histograma(imagen)
```
