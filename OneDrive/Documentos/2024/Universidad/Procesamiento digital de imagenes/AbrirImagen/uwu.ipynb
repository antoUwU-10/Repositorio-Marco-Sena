```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image =  cv2.imread('imagen.jpg', cv2.IMREAD_GRAYSCALE)

original_img = image.dtype 
print('Profundida original:', original_img)

image_16bits = np.uint16(image)

nueva_img = image_16bits.dtype
print('Profundida original:', nueva_img)

cv2.imwrite('image16bits.jpg', image_16bits)

img_guardada = cv2.imread('image16bits.jpg', cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.title('Original')
plt.imshow(image, cmap = 'gray')
plt.axis('off')

plt.subplot(122)
plt.title('Imagen tratada 16 Bits')
plt.imshow(img_guardada, cmap = 'gray')
plt.axis('off')
```
