```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [10,5]

def open_img(img,selec):
    #Agarrar las imagenes convertidas
    image = cv2.imread(img, cv2.COLOR_RGB2BGR)

    #Pa guardar la imagen
    cv2.imwrite('imagenModificada.jpg', cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))

    #Pa mostrar las imgenes tuki tuki
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis(selec)

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='gray')
    plt.axis(selec)

open_img("wallpaperbetter (1).jpg", 'off')
```
